# *Multimodal Data Generator*

```python
import os
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder
import numpy as np

class MultimodalDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, dataframe, batch_size=32, img_size=(299, 299), num_classes=4, shuffle=True):
        super().__init__()
        self.df = dataframe.copy()
        self.batch_size = batch_size
        self.img_size = img_size
        self.num_classes = num_classes
        self.shuffle = shuffle
        self.indexes = np.arange(len(self.df))

        # Encode labels
        self.label_encoder = LabelEncoder()
        self.df['label_encoded'] = self.label_encoder.fit_transform(self.df['label'])

        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.df) / self.batch_size))

    def __getitem__(self, index):
        batch_indexes = self.indexes[index * self.batch_size:(index + 1) * self.batch_size]
        batch_df = self.df.iloc[batch_indexes]

        # ‚úÖ **Ensure 1000 features are selected**
        cm_features = batch_df.drop(columns=['spot_id', 'img', 'label', 'label_encoded', 'X', 'Y']).values.astype('float32')

        # ‚úÖ **Check count matrix shape**
        if cm_features.shape[1] != 1000:
            raise ValueError(f"‚ùå Count matrix has {cm_features.shape[1]} features instead of 1000.")

        images = np.array([self.safe_load_img(row['img']) for _, row in batch_df.iterrows()])
        batch_labels = batch_df['label_encoded'].values
        labels = to_categorical(batch_labels, num_classes=self.num_classes)

        # ‚úÖ **Convert to TensorFlow Tensors**
        images = tf.convert_to_tensor(images, dtype=tf.float32)
        cm_features = tf.convert_to_tensor(cm_features, dtype=tf.float32)
        labels = tf.convert_to_tensor(labels, dtype=tf.float32)

        return (images, cm_features), labels

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def safe_load_img(self, path):
        """ Load image safely, return blank if missing """
        try:
            return img_to_array(load_img(path, target_size=self.img_size)) / 255.0
        except Exception as e:
            print(f"‚ö†Ô∏è Warning: Could not load image {path}. Error: {e}")
            return np.zeros((*self.img_size, 3))  # Return blank image
```

```python
# Train-Validation-Test split (Stratified)
train_df, test_df = train_test_split(combined_df, test_size=0.2, stratify=combined_df['label'], random_state=42)
train_df, val_df = train_test_split(train_df, test_size=0.2, stratify=train_df['label'], random_state=42)

print("Train Labels:\n", train_df['label'].value_counts())
print("Validation Labels:\n", val_df['label'].value_counts())
print("Test Labels:\n", test_df['label'].value_counts())

train_gen = MultimodalDataGenerator(train_df, batch_size=32, num_classes=4)
val_gen = MultimodalDataGenerator(val_df, batch_size=32, num_classes=4)
test_gen = MultimodalDataGenerator(test_df, batch_size=32, num_classes=4)
```

**üîπ Step 3: Define the Multimodal Model**

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.regularizers import l2

# ‚úÖ ResNet model for Image Feature Extraction
def build_resnet(input_shape):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    base_model.trainable = False  # Freeze ResNet layers
    inputs = layers.Input(shape=input_shape)
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)  # Convert to 1D vector
    x = layers.Dense(128, activation='relu')(x)  # Projection Layer
    return Model(inputs, x, name="ResNet_Encoder")

# ‚úÖ MLP model for Gene Expression Feature Extraction
def build_mlp(input_dim):
    inputs = layers.Input(shape=(input_dim,))
    x = layers.Dense(128, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    return Model(inputs, x, name="MLP_Encoder")

# ‚úÖ Final Multimodal Fusion Model
def build_multimodal_model(img_shape, gene_dim, num_classes):
    # Feature Extractors
    resnet_encoder = build_resnet(img_shape)
    mlp_encoder = build_mlp(gene_dim)

    # Inputs
    img_input = layers.Input(shape=img_shape, name="Image_Input")
    gene_input = layers.Input(shape=(gene_dim,), name="Gene_Input")

    # Feature Extraction
    histology_feat = resnet_encoder(img_input)
    gene_feat = mlp_encoder(gene_input)

    # ‚úÖ **Fusion via Concatenation (No cross-attention)**
    fused = layers.Concatenate()([histology_feat, gene_feat])
    
    # Fully Connected Layers
    fused = layers.Dense(128, activation='relu')(fused)
    fused = layers.Dense(64, activation='relu')(fused)

    # **Final Classification Layer**
    output = layers.Dense(num_classes, activation='softmax', name="Predicted_Label")(fused)

    # ‚úÖ **Model Definition (No Label Input Layer)**
    model = Model(inputs=[img_input, gene_input], outputs=output, name="CrossModalModel")
    
    return model

# ‚úÖ Example Model Initialization
img_shape = (299, 299, 3)  # Image shape for ResNet
gene_dim = 1000  # Gene Feature Size (Ensure matches your dataset)
num_classes = 4  # Disease Stages

# Build the model
model = build_multimodal_model(img_shape, gene_dim, num_classes)
model.summary()

# ‚úÖ **Compile the Model** (with AUC metric)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy", tf.keras.metrics.AUC(name='auc')]  # Add AUC here
)
```

**üîπ Step 4: Train and Evaluation**

```python
# Sample batch from training generator
sample_train_batch = train_gen[0]

# Print batch structure
print("Training batch structure:")
print(f"Image shape: {sample_train_batch[0][0].shape}")  # (batch_size, 299, 299, 3)
print(f"Count matrix shape: {sample_train_batch[0][1].shape}")  # (batch_size, num_genes)
print(f"Labels shape: {sample_train_batch[1].shape}")  # (batch_size, num_classes)
```

```python

# Define Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_accuracy', 
        patience=5, 
        restore_best_weights=True
    )
]

# ‚úÖ **Train Model**
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=50,
    callbacks=callbacks
)


```

```python
def sensitivity_specificity_multiclass(y_true, y_pred, labels=None):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    n_classes = cm.shape[0]
    
    sensitivity = []
    specificity = []

    for i in range(n_classes):
        TP = cm[i, i]
        FN = cm[i, :].sum() - TP
        FP = cm[:, i].sum() - TP
        TN = cm.sum() - (TP + FN + FP)

        sensitivity.append(TP / (TP + FN) if (TP + FN) > 0 else 0.0)
        specificity.append(TN / (TN + FP) if (TN + FP) > 0 else 0.0)

    return np.array(sensitivity), np.array(specificity)

sensitivity, specificity = sensitivity_specificity_multiclass(y_true, y_pred)

for i, (sens, spec) in enumerate(zip(sensitivity, specificity)):
    print(f"Class {i}: Sensitivity = {sens:.2f}, Specificity = {spec:.2f}")
```
